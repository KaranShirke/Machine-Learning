```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import math

titanic_data = pd.read_csv('titanic.csv')
#showing first 10 values from dataset
titanic_data.head(10)
```

```python
titanic_data.drop("Cabin", axis=1, inplace=True)
```

```python
titanic_data.head(10)
```

```python
titanic_data.dropna(inplace=True)
```

```python
titanic_data.head(10)
```

```python
sex = pd.get_dummies(titanic_data['Sex'],prefix="Sex", prefix_sep="_")
```

```python
sex.head()
```

```python
pclass = pd.get_dummies(titanic_data['Pclass'], prefix="Pclass", prefix_sep="_")
```

```python
pclass.head()
```

```python
sibsp = pd.get_dummies(titanic_data['SibSp'], prefix="SibSp", prefix_sep="_")
```

```python
sibsp.head()
```

```python
parch = pd.get_dummies(titanic_data['Parch'], prefix="Parch", prefix_sep="_")
```

```python
parch.head()
```

```python
embarked = pd.get_dummies(titanic_data['Embarked'], prefix="Embarked", prefix_sep="_")
```

```python
embarked.head()
```

```python
titanic_data=pd.concat([titanic_data,sex,pclass,sibsp,parch,embarked], axis=1)
```

```python
titanic_data.head(3)
```

```python

titanic_data.drop(['PassengerId'], axis=1,inplace=True)
```

```python
titanic_data.isnull().sum()
```

```python
titanic_data.head(10)
```

```python
y = titanic_data['Survived']
X = titanic_data.drop(['Survived'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 101, stratify=y, test_size=0.25)
```

```python
y_train.value_counts(normalize=True)
```

```python
y_test.value_counts(normalize=True)
```

```python
X_train.shape, y_train.shape
```

```python
X_test.shape, y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import DecisionTreeRegressor
```

```python
dt_model = DecisionTreeClassifier(random_state=10)
```

```python
dt_model.fit(X_train, y_train)
```

```python
dt_model.score(X_train, y_train)*100
```

```python
dt_model.score(X_test, y_test)*100
```

```python
dt_model.predict(X_test)
```

```python
dt_model.predict_proba(X_test)
```

 ## Changing the max_depth

```python
train_accuracy = []
validation_accuracy = []
for depth in range(1,10):
    dt_model = DecisionTreeClassifier(max_depth=depth, random_state=10)
    dt_model.fit(X_train, y_train)
    train_accuracy.append(dt_model.score(X_train, y_train))
    validation_accuracy.append(dt_model.score(X_test, y_test))
```

```python
frame = pd.DataFrame({'max_depth':range(1,10), 'train_acc':train_accuracy, 'valid_acc':validation_accuracy})
frame.head()
```

```python
plt.figure(figsize=(12,6))
plt.plot(frame['max_depth'], frame['train_acc'], marker='o')
plt.plot(frame['max_depth'], frame['valid_acc'], marker='o')
plt.xlabel('Depth of tree')
plt.ylabel('performance')
plt.legend()
```

```python
dt_model = DecisionTreeClassifier(max_depth=4, max_leaf_nodes=25, random_state=10)
```

```python
dt_model.fit(X_train, y_train)
```

```python
dt_model.score(X_train, y_train)
```
